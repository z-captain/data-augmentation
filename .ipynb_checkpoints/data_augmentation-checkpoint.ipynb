### CNN architectures for image classification and  data augmentation technique

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread # matplotlib funtion that is used to load RGB pictured

def load_data(img_path, label_path):
    """function that loads the data """
    with open(label_path) as file:
        class_labels = [label.strip("\n").split('\t') for label in file.readlines()]
    
    images = []
    labels = []
    for file, label in class_labels:
        images.append(imread(os.path.join(img_path, file)))
        labels.append(int(label))
    return np.array(images), np.array(labels)

train_images, train_labels = load_data('original_data/train', 'original_data/train_class_labels.txt')
test_images, test_labels = load_data('original_data/test', 'original_data/test_class_labels.txt')

train_images_augmented, train_labels_augmented =\
        load_data('augmented_data/train', 'augmented_data/train_class_labels.txt')

# print shapes 
print("training size:", train_images.shape)
print("training class labels size:", train_labels.shape)

print("\ntest size:", test_images.shape)
print("test class labels size:", test_labels.shape)
```

```python
from keras.models import Sequential # what kind of model ? a sequenctial model
from keras.layers.core import Dense, Dropout # different layers, activation function, and dropout
from keras.optimizers import Adam # optimization algorithm
from keras.layers.normalization import BatchNormalization
from keras.layers.convolutional import Conv2D
from keras.layers.convolutional import MaxPooling2D
from keras.layers.core import Flatten
from keras.losses import binary_crossentropy

def build_model(lr, show_summary=False):
    """ function that builds a keras model using the provided hyperparameters """
    model = Sequential(name=f"CNN model - Learning_rate: {lr}")
    model.add(
        Conv2D(32, kernel_size = (3,3), activation="relu", input_shape = (32,32,3))
    )
    model.add(
        MaxPooling2D(pool_size=(2,2), )
    )
    model.add(BatchNormalization())
    model.add(Dropout(0.2))
    model.add(
        Conv2D(64, kernel_size = (3,3), activation="relu")
    )
    model.add(
        MaxPooling2D(pool_size=(2,2), )
    )
    model.add(BatchNormalization())
    model.add(Dropout(0.3))
    model.add(Flatten())
    model.add(
        Dense(100, activation="relu")
    )
    model.add(BatchNormalization())
    model.add(Dropout(0.5))
    model.add(
        Dense(1, activation="sigmoid")
    )
    
    optimizer = Adam(learning_rate=lr)
    model.compile(loss= binary_crossentropy, optimizer=optimizer, metrics=["accuracy"])
    if show_summary:
        model.summary()
    
    return model
```

```python
# train the model 

NUM_EPOCHS = 40

cnn_model_1 = build_model(0.05, show_summary=True)

history_1 = cnn_model_1.fit(
    train_images, 
    train_labels, 
    validation_data = [test_images, test_labels],
    epochs=NUM_EPOCHS,
    shuffle=True, 
    verbose=0
)
```

```python
cnn_model_2 = build_model(0.001)

history_2 = cnn_model_2.fit(
    train_images, 
    train_labels, 
    validation_data = [test_images, test_labels],
    epochs=NUM_EPOCHS,
    shuffle=True,
    verbose=0
)
```

```python
cnn_model_3 = build_model(0.0005)

history_3 = cnn_model_3.fit(
    train_images, 
    train_labels, 
    validation_data = [test_images, test_labels],
    epochs=NUM_EPOCHS,
    shuffle=True,
    verbose=0
)
```

```python
cnn_model_augmented_1 = build_model(0.05)

history_augmented_1 = cnn_model_augmented_1.fit(
    train_images_augmented, 
    train_labels_augmented, 
    validation_data = [test_images, test_labels],
    epochs=NUM_EPOCHS,
    shuffle=True,
    verbose=0
)
```

```python
cnn_model_augmented_2 = build_model(0.001)

history_augmented_2 = cnn_model_augmented_2.fit(
    train_images_augmented, 
    train_labels_augmented, 
    validation_data = [test_images, test_labels],
    epochs=NUM_EPOCHS,
    shuffle=True,
    verbose=0
)
```

```python
cnn_model_augmented_3 = build_model(0.0005)

history_augmented_3 = cnn_model_augmented_3.fit(
    train_images_augmented, 
    train_labels_augmented, 
    validation_data = [test_images, test_labels],
    epochs=NUM_EPOCHS,
    shuffle=True,
    verbose=0
)
```

```python
def plotHistory(Tuning):
    fig, ax = plt.subplots(1,2,figsize=(15,5))
    ax[0].plot(Tuning.history['loss'])
    ax[0].plot(Tuning.history['val_loss'])
    ax[0].set_title('Loss vs Epoch')
    ax[0].set_ylabel('Loss (binary crossentropy)')
    ax[0].set_xlabel('Epoch')
    ax[0].legend(['Train', 'Validation'], loc='best')
    
    ax[1].plot(Tuning.history['accuracy'])
    ax[1].plot(Tuning.history['val_accuracy'])
    ax[1].set_title('Accuracy vs Epoch')
    ax[1].set_ylabel('Accuracy')
    ax[1].set_xlabel('Epoch')
    ax[1].set_ylim([0.0,1.0])
    ax[1].legend(['Train', 'Validation'], loc='best')
    plt.show(block = False)
    plt.show()
```

```python
# the best learning rate for the model trained on original data: 0.005

plotHistory(history_3)
```

```python
# the best learning rate for the model trained on augmented data: 0.005

plotHistory(history_augmented_3)
```
